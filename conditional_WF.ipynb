```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict,Literal
```

```python
class Quadratic_Eq(TypedDict):
    a:str
    b:str
    c:str
    equation:str
    discriminant:float
    result:str
    
```

```python
graph=StateGraph(Quadratic_Eq)
```

```python
def show_Eq(state:Quadratic_Eq)->Quadratic_Eq:
    eq=f'{state['a']}x2+{state['b']}x+{state["c"]}'
    return {'equation':eq}
```

```python
def calculate_dis(state:Quadratic_Eq)->Quadratic_Eq:
    dis=(state['b']**2)-(4*(state['a']*state['c']))
    return {'discriminant':dis}
```

```python
def real_roots(state:Quadratic_Eq):
    root1=((-state['b']+state['discriminant']**0.5)/(2*state['a']))
    root2=((-state['b']-state['discriminant']**0.5)/(2*state['a']))
    res=f'results are root1:{root1} and root2:{root2}'
    return {'result':res}
```

```python
def repeated_roots(state:Quadratic_Eq):
    root=(-state['b']/(2*state['a']))
    
    res=f'only reprated root is {root}'
    return {'result':res}
```

```python
def no_real_roots(state:Quadratic_Eq):
    res=f'no real root present'
    return {'result':res}
```

```python
def check_condition(state:Quadratic_Eq)->Literal["real_roots","repeated_roots","no_real_roots"]:
    if state['discriminant']>0:
        return "real_roots"
    elif state['discriminant']==0:
        return "repeated_roots"
    else:
        return no_real_roots
```

```python
graph.add_node('show_Equation',show_Eq)
graph.add_node('calculate_discriminant',calculate_dis)
graph.add_node('real_roots',real_roots)
graph.add_node('repeated_roots',repeated_roots)
graph.add_node('no_real_roots',no_real_roots)

```

```python
graph.add_edge(START,'show_Equation')
graph.add_edge('show_Equation','calculate_discriminant')
graph.add_conditional_edges('calculate_discriminant',check_condition)
graph.add_edge('no_real_roots',END)
graph.add_edge('real_roots',END)
graph.add_edge('repeated_roots',END)
```

```python
wf=graph.compile()
```

```python
intial_state={
    'a':4,'b':-5,'c':-4
}
wf.invoke(intial_state)
```
