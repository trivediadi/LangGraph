```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class Criket(TypedDict):
    runs:int 
    balls:int
    fours:int
    sixes:int
    sr:float
    run_pb:float
    ball_pb:float
    summary=str
```

```python
graph=StateGraph(Criket)
```

```python
def sr(state:Criket)->Criket:
    run=state['runs']
    balls=state['balls']
    sr=(run/balls)*100
    
    return {'sr':sr}
```

```python
def rpb(state:Criket)->Criket:
    four=state['fours']
    six=state['sixes']
    total=(six*6+four*4)/(state['runs'])*100
    return {'run_pb':total}
```

```python
def bpb(state:Criket)->Criket:
    four=state['fours']
    six=state['sixes']
    total=(state['balls'])/(four+six)
    return {'ball_pb':total}
```

```python
def summary(state:Criket)->Criket:
    summ=f"""
Strike Rate- {state['sr']}
Balls per boundary-{state['ball_pb']}
Runs per boundary-{state['run_pb']}
"""
    state['summary']=summ
    return state
```

```python
graph.add_node('strike_rate',sr)
graph.add_node('runs_per_boundary',rpb)
graph.add_node('ball_per_boundary',bpb)
```

```python
graph.add_node('summary',summary)
```

```python
graph.add_edge(START,'strike_rate')
graph.add_edge(START,'runs_per_boundary')
graph.add_edge(START,'ball_per_boundary')
graph.add_edge('strike_rate','summary')
graph.add_edge('runs_per_boundary','summary')
graph.add_edge('ball_per_boundary','summary')
graph.add_edge('summary',END)
```

```python
wf=graph.compile()
```

```python
intial_state={'runs':100,
    'balls':50,
    'fours':6,
    'sixes':4
    }
wf.invoke(intial_state)
```
