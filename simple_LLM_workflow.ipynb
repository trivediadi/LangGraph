```python
from langgraph.graph import StateGraph,START,END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatGoogleGenerativeAI(model="gemini-2.0-flash")
```

```python
# create a state  
class LLMState(TypedDict):
    question:str
    answer:str
```

```python
def qamodel(state:LLMState)->LLMState:
    #extract the question from state
    question=state['question']
    result=model.invoke(question)
    state['answer']=result.content
    return state
```

```python
#create a graph
graph=StateGraph(LLMState)

#create node
graph.add_node('qamodel',qamodel)
#create edge
graph.add_edge(START,'qamodel')
graph.add_edge('qamodel',END)
#compile
workflow=graph.compile()
```

```python
intial_workflow={'question':"What is the color of sun?"}
result=workflow.invoke(intial_workflow)
```
